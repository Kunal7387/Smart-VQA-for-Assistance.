```python
import json
from collections import Counter
import matplotlib.pyplot as plt
from wordcloud import WordCloud
import seaborn as sns
from itertools import combinations
import pandas as pd

# File paths
train_json = "/Users/krutikotadia/Desktop/DLP Project/Annotations/filtered_train.json"
val_json = "/Users/krutikotadia/Desktop/DLP Project/Annotations/filtered_val.json"
test_json = "/Users/krutikotadia/Desktop/DLP Project/Annotations/filtered_test.json"

def load_json(file_path):
    """Load JSON file."""
    with open(file_path, 'r') as f:
        return json.load(f)

def analyze_questions(data):
    """Analyze question statistics."""
    questions = [entry['question'] for entry in data]
    lengths = [len(q.split()) for q in questions]
    avg_length = sum(lengths) / len(lengths)
    max_length = max(lengths)
    min_length = min(lengths)

    print(f"Total Questions: {len(questions)}")
    print(f"Average Question Length: {avg_length:.2f} words")
    print(f"Max Question Length: {max_length} words")
    print(f"Min Question Length: {min_length} words")

    # Plot question length distribution
    plt.hist(lengths, bins=20, alpha=0.7, color="blue", edgecolor="black")
    plt.title("Question Length Distribution")
    plt.xlabel("Length (words)")
    plt.ylabel("Frequency")
    plt.show()

    # Generate a word cloud of questions
    question_text = " ".join(questions)
    wordcloud = WordCloud(width=800, height=400, background_color="white").generate(question_text)
    plt.figure(figsize=(10, 5))
    plt.imshow(wordcloud, interpolation="bilinear")
    plt.axis("off")
    plt.title("Word Cloud of Questions")
    plt.show()

    return questions

def analyze_answers(data):
    """Analyze answer statistics."""
    if 'answer' in data[0]:  # Check if answer key exists
        answers = [entry['answer'] for entry in data]
        answer_counts = Counter(answers)

        print(f"Total Unique Answers: {len(answer_counts)}")
        print("Top 10 Answers:")
        for answer, count in answer_counts.most_common(10):
            print(f"{answer}: {count}")

        # Plot top 10 answers
        top_answers = answer_counts.most_common(10)
        labels, counts = zip(*top_answers)
        plt.bar(labels, counts, alpha=0.7, color="green", edgecolor="black")
        plt.title("Top 10 Answers")
        plt.xlabel("Answer")
        plt.ylabel("Frequency")
        plt.xticks(rotation=45)
        plt.show()
    else:
        print("No 'answer' key in this dataset.")

def word_cooccurrence_heatmap(questions):
    """Generate a heatmap of word co-occurrences in questions."""
    # Tokenize the questions
    tokenized_questions = [q.split() for q in questions]

    # Build co-occurrence counts
    cooccurrence_counts = Counter()
    for tokens in tokenized_questions:
        for pair in combinations(sorted(set(tokens)), 2):  # Unique pairs
            cooccurrence_counts[pair] += 1

    # Prepare data for heatmap
    words = list({word for pair in cooccurrence_counts.keys() for word in pair})
    word_to_idx = {word: i for i, word in enumerate(words)}
    cooccurrence_matrix = [[0] * len(words) for _ in range(len(words))]
    for (w1, w2), count in cooccurrence_counts.items():
        i, j = word_to_idx[w1], word_to_idx[w2]
        cooccurrence_matrix[i][j] = count
        cooccurrence_matrix[j][i] = count  # Symmetric

    # Convert to DataFrame for heatmap
    cooccurrence_df = pd.DataFrame(cooccurrence_matrix, index=words, columns=words)

    # Plot heatmap
    plt.figure(figsize=(12, 10))
    sns.heatmap(cooccurrence_df, cmap="Blues", square=True, cbar=True)
    plt.title("Word Co-occurrence Heatmap")
    plt.show()

def analyze_answer_types(data):
    """Analyze answer type statistics (for val.json)."""
    if 'answer_type' in data[0]:  # Check if answer_type key exists
        answer_types = [entry['answer_type'] for entry in data]
        answer_type_counts = Counter(answer_types)

        print(f"Answer Type Distribution:")
        for a_type, count in answer_type_counts.items():
            print(f"{a_type}: {count}")

        # Plot answer type distribution
        labels, counts = zip(*answer_type_counts.items())
        plt.pie(counts, labels=labels, autopct='%1.1f%%', startangle=140)
        plt.title("Answer Type Distribution")
        plt.show()
    else:
        print("No 'answer_type' key in this dataset.")

# Load data
train_data = load_json(train_json)
val_data = load_json(val_json)
test_data = load_json(test_json)

# Analyze train dataset
print("\n--- Train Dataset ---")
train_questions = analyze_questions(train_data)
analyze_answers(train_data)

# Analyze val dataset
print("\n--- Validation Dataset ---")
val_questions = analyze_questions(val_data)
analyze_answers(val_data)
analyze_answer_types(val_data)  # Specific to val.json

# Analyze test dataset
print("\n--- Test Dataset ---")
test_questions = analyze_questions(test_data)

# Generate a word co-occurrence heatmap for train questions
word_cooccurrence_heatmap(train_questions)
```

### Data Preprocessing

```python
import os
import json
from PIL import Image
import numpy as np
from transformers import BertTokenizer
from sklearn.preprocessing import LabelEncoder
import torch
from torchvision import transforms

# Paths (Update these accordingly)
train_images_dir = "/Users/krutikotadia/Desktop/DLP Project/train/images_train"
val_images_dir = "/Users/krutikotadia/Desktop/DLP Project/val/images_val"
val_json_file = "/Users/krutikotadia/Desktop/DLP Project/Annotations/filtered_val.json"

# Constants
MAX_SEQ_LEN = 30  # Max tokens for questions
IMG_SIZE = (224, 224)  # Image dimensions for resizing

# Initialize tokenizer and label encoder
tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')
label_encoder = LabelEncoder()

# Image transformations
image_transforms = transforms.Compose([
    transforms.Resize(IMG_SIZE),
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])  # ImageNet normalization
])

def load_json(file_path):
    """Load JSON file."""
    with open(file_path, 'r') as f:
        return json.load(f)

def preprocess_question(question):
    """
    Tokenize the question and truncate/pad to MAX_SEQ_LEN.
    Returns input_ids and attention_mask.
    """
    encoded = tokenizer(
        question,
        max_length=MAX_SEQ_LEN,
        padding="max_length",
        truncation=True,
        return_tensors="pt"
    )
    return encoded["input_ids"].squeeze(0), encoded["attention_mask"].squeeze(0)

def preprocess_image(image_path):
    """
    Load and preprocess the image.
    Returns a normalized tensor.
    """
    image = Image.open(image_path).convert("RGB")
    return image_transforms(image)

def preprocess_answers(data):
    """
    Extract all unique answers from the training data and encode them.
    Returns the encoded answers and a LabelEncoder object.
    """
    answers = [entry["answer"] for entry in data if "answer" in entry]
    label_encoder.fit(answers)  # Fit the encoder
    encoded_answers = label_encoder.transform(answers)
    return encoded_answers

def preprocess_data(data, images_dir):
    """
    Preprocess the dataset.
    Returns preprocessed questions (input_ids, attention_mask), images, and answers.
    """
    questions_input_ids = []
    questions_attention_masks = []
    images = []
    answers = []

    for entry in data:
        # Process question
        input_ids, attention_mask = preprocess_question(entry["question"])
        questions_input_ids.append(input_ids)
        questions_attention_masks.append(attention_mask)

        # Process image
        image_path = os.path.join(images_dir, entry["image"])
        if os.path.exists(image_path):
            images.append(preprocess_image(image_path))
        else:
            print(f"Image {entry['image']} not found! Skipping.")

        # Process answer (if available)
        if "answer" in entry:
            answers.append(entry["answer"])

    # Convert to tensors
    questions_input_ids = torch.stack(questions_input_ids)
    questions_attention_masks = torch.stack(questions_attention_masks)
    images = torch.stack(images)

    # Encode answers if available
    encoded_answers = None
    if answers:
        encoded_answers = label_encoder.transform(answers)
        encoded_answers = torch.tensor(encoded_answers)

    return questions_input_ids, questions_attention_masks, images, encoded_answers

# Load data
train_data = load_json(train_json)
val_data = load_json(val_json)

# Preprocess training data
print("Preprocessing training data...")
train_encoded_answers = preprocess_answers(train_data)
train_input_ids, train_attention_masks, train_images, train_labels = preprocess_data(train_data, train_images_dir)

# Preprocess validation data
print("Preprocessing validation data...")
val_input_ids, val_attention_masks, val_images, val_labels = preprocess_data(val_data, val_images_dir)

# Save LabelEncoder for later use
with open("label_encoder.json", "w") as f:
    json.dump(label_encoder.classes_.tolist(), f)

print("Data preprocessing completed.")
```

### Model Building

```python
import torch
import torch.nn as nn
from torchvision.models import resnet18
from transformers import BertModel

class BaselineVQAModel(nn.Module):
    def __init__(self, num_classes):
        super(BaselineVQAModel, self).__init__()
        # Pretrained ResNet for image feature extraction
        self.image_model = resnet18(pretrained=True)
        self.image_model.fc = nn.Identity()  # Remove the final classification layer
        
        # Pretrained BERT for text feature extraction
        self.text_model = BertModel.from_pretrained('bert-base-uncased')
        
        # Combine image and text features
        self.fc = nn.Sequential(
            nn.Linear(512 + 768, 512),
            nn.ReLU(),
            nn.Linear(512, num_classes)
        )
    
    def forward(self, input_ids, attention_mask, images):
        # Extract image features
        image_features = self.image_model(images)  # Shape: [batch_size, 512]
        
        # Extract text features
        text_outputs = self.text_model(input_ids=input_ids, attention_mask=attention_mask)
        text_features = text_outputs.last_hidden_state[:, 0, :]  # [CLS] token embeddings
        
        # Concatenate image and text features
        combined_features = torch.cat((image_features, text_features), dim=1)
        
        # Classify combined features
        output = self.fc(combined_features)
        return output
```

```python
import numpy as np

# Extend the LabelEncoder
def extend_label_encoder(label_encoder, additional_labels):
    """
    Extends a fitted LabelEncoder with new labels, ensuring classes_ is a NumPy array.
    """
    current_classes = set(label_encoder.classes_)
    new_classes = set(additional_labels)
    combined_classes = list(current_classes | new_classes)
    combined_classes.sort()  # Keep classes sorted
    label_encoder.classes_ = np.array(combined_classes)  # Convert to NumPy array
    return label_encoder


# Extract all unique answers from the validation dataset
def extract_all_unique_answers(json_file):
    """
    Extracts all unique answers from the validation JSON.
    """
    with open(json_file, 'r') as f:
        data = json.load(f)
    all_answers = []
    for entry in data:
        valid_answers = [ans['answer'] for ans in entry['answers'] if ans['answer_confidence'] in ["yes", "maybe"]]
        all_answers.extend(valid_answers)
    return set(all_answers)

# Extract unique validation answers
validation_unique_answers = extract_all_unique_answers(val_json_file)

# Extend the LabelEncoder with validation labels
label_encoder = extend_label_encoder(label_encoder, validation_unique_answers)

print("LabelEncoder extended with validation labels.")

# Preprocess validation data again
val_input_ids, val_attention_masks, val_images, val_labels = preprocess_validation_data(val_json_file, val_image_dir)

print("Validation preprocessing completed.")
```

### Building the attention mechanism

```python
import torch.nn.functional as F

class AttentionMechanism(nn.Module):
    def __init__(self, image_dim, text_dim, hidden_dim):
        super(AttentionMechanism, self).__init__()
        self.image_proj = nn.Linear(image_dim, hidden_dim)
        self.text_proj = nn.Linear(text_dim, hidden_dim)
        self.attention_proj = nn.Linear(hidden_dim, 1)

    def forward(self, image_features, text_features):
        """
        Compute attention scores and apply them to image features.
        Args:
            image_features: Tensor of shape [batch_size, num_regions, image_dim]
            text_features: Tensor of shape [batch_size, text_dim]
        Returns:
            attended_image_features: Tensor of shape [batch_size, image_dim]
        """
        # Project image and text features
        image_proj = self.image_proj(image_features)  # [batch_size, num_regions, hidden_dim]
        text_proj = self.text_proj(text_features).unsqueeze(1)  # [batch_size, 1, hidden_dim]

        # Compute attention scores
        combined_features = torch.tanh(image_proj + text_proj)  # [batch_size, num_regions, hidden_dim]
        attention_scores = self.attention_proj(combined_features).squeeze(-1)  # [batch_size, num_regions]

        # Apply softmax to get attention weights
        attention_weights = F.softmax(attention_scores, dim=1)  # [batch_size, num_regions]

        # Compute attended image features
        attended_image_features = torch.sum(image_features * attention_weights.unsqueeze(-1), dim=1)  # [batch_size, image_dim]

        return attended_image_features
```

```python
class AttentionVQAModel(nn.Module):
    def __init__(self, num_classes):
        super(AttentionVQAModel, self).__init__()
        # Pretrained ResNet for image feature extraction
        self.image_model = resnet18(pretrained=True)
        self.image_model.fc = nn.Identity()  # Remove the final classification layer

        # Pretrained BERT for text feature extraction
        self.text_model = BertModel.from_pretrained('bert-base-uncased')

        # Attention mechanism
        self.attention = AttentionMechanism(image_dim=512, text_dim=768, hidden_dim=256)

        # Fully connected layers
        self.fc = nn.Sequential(
            nn.Linear(512 + 768, 512),
            nn.ReLU(),
            nn.Dropout(0.5),
            nn.Linear(512, num_classes)
        )
    
    def forward(self, input_ids, attention_mask, images):
        # Extract image features
        image_features = self.image_model(images)  # Shape: [batch_size, num_regions, 512]
        
        # Extract text features
        text_outputs = self.text_model(input_ids=input_ids, attention_mask=attention_mask)
        text_features = text_outputs.last_hidden_state[:, 0, :]  # [CLS] token embeddings

        # Apply attention
        attended_image_features = self.attention(image_features, text_features)

        # Concatenate attended image features and text features
        combined_features = torch.cat((attended_image_features, text_features), dim=1)

        # Classify combined features
        output = self.fc(combined_features)
        return output
```
